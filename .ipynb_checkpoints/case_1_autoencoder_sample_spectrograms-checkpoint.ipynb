```python
#code from https://ramhiser.com/post/2018-05-14-autoencoders-with-keras/

%reset -f 

from IPython.display import Image, SVG
import matplotlib.pyplot as plt

plt.style.use('../misc/demo_style.mplstyle')

import numpy as np
import keras
from keras.datasets import mnist
from keras.models import Model, Sequential
from keras.layers import Input, Dense, Conv2D, MaxPooling2D, UpSampling2D, Flatten, Reshape
from keras import regularizers
from sklearn.model_selection import train_test_split
```

```python
# Loads the training and test data sets (ignoring class labels)
#(x_train, _), (x_test, _) = mnist.load_data()

#want our scalograms instead of mnist

#also q3 and q67 i think
#are the data processed in order? 
#also need to store the timestamps for each image
#where did these come from
#process_scalograms_data.ipynb in data_prep folder in repo

q001_data_raw = np.load('/data0/stephey/processed_data/q001_data_raw_cut.npy')
q003_data_raw = np.load('/data0/stephey/processed_data/q003_data_raw_cut.npy')
q076_data_raw = np.load('/data0/stephey/processed_data/q076_data_raw_cut.npy')
q103_data_raw = np.load('/data0/stephey/processed_data/q103_data_raw_cut.npy')

##this is all data, take only bump to start (0 is stft, 1 is bump, 2 is morse)
#q001_data_raw=q001_data_raw[:,:,:,1]
#q003_data_raw=q003_data_raw[:,:,:,1]
#q076_data_raw=q076_data_raw[:,:,:,1]
#q103_data_raw=q103_data_raw[:,:,:,1]

#don't normalize over the whole quench, normalize each image
q001_data = np.zeros_like(q001_data_raw)
q003_data = np.zeros_like(q003_data_raw)
q076_data = np.zeros_like(q076_data_raw)
q103_data = np.zeros_like(q103_data_raw)

#it's not pretty but it works
for i, im in enumerate(q001_data_raw):
    q001_data[i,:,:] = q001_data_raw[i,:,:]/np.max(q001_data_raw[i,:,:])

for i, im in enumerate(q003_data_raw):
    q003_data[i,:,:] = q003_data_raw[i,:,:]/np.max(q003_data_raw[i,:,:])
    
for i, im in enumerate(q076_data_raw):
    q076_data[i,:,:] = q076_data_raw[i,:,:]/np.max(q076_data_raw[i,:,:])
    
for i, im in enumerate(q103_data_raw):
    q103_data[i,:,:] = q103_data_raw[i,:,:]/np.max(q103_data_raw[i,:,:])    
    
#we shuffle in the training so it's ok
qall_data = np.concatenate((q001_data, q003_data, q076_data, q103_data), axis=0)

print(qall_data.shape)    
```

```python
from keras.models import load_model



#load our previously trained model
autoencoder = load_model('case_1_autoencoder')

encoder = Model(inputs=autoencoder.input, outputs=autoencoder.get_layer('dense_3').output)
encoder.summary()
```

```python
#now put all our data in the encoder why not

qall_reshape = qall_data.reshape((len(qall_data), np.prod(qall_data.shape[1:])))
encoded_all = encoder.predict(qall_reshape)

print(encoded_all.shape)
```

```python
#Try pca on encoded images why not
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
#need to flatten i guess?
img_pca = pca.fit_transform(encoded_all)

#boundary 1
boundary1 = q001_data.shape[0]
#boundary 2
boundary2 = boundary1 + q003_data.shape[0]
#boundary 3
boundary3 = boundary2 + q076_data.shape[0]

color_key=['red','orange','green','blue']
color_label=[]
for i in range(qall_data.shape[0]):
    if i < boundary1:
        color_label.append(color_key[0])
    elif boundary1 < i <= boundary2:
        color_label.append(color_key[1])
    elif boundary2 < i <= boundary3: 
        color_label.append(color_key[2])
    else:
        color_label.append(color_key[3])
        
print(len(color_label))   
print(len(img_pca))
```

```python
#try making a nicer figure for maxim

from matplotlib.patches import Patch
from matplotlib.lines import Line2D

msize = 16

#hack the legend together
legend_elements = [Line2D([0], [0], marker='o', color='w', label='Quench 001',
                          markerfacecolor='red', markersize=8),
                   Line2D([0], [0], marker='o', color='w', label='Quench 003',
                          markerfacecolor='orange', markersize=8),
                   Line2D([0], [0], marker='o', color='w', label='Quench 076',
                          markerfacecolor='green', markersize=8),
                   Line2D([0], [0], marker='o', color='w', label='Quench 103',
                          markerfacecolor='blue', markersize=8)]

# Create the figure and add our hacked legend
plt.figure(figsize=(10, 7))
for i in range(qall_data.shape[0]):
    plt.scatter(img_pca[i,0],img_pca[i,1], color=color_label[i], s=msize)
plt.title('Case 1 PCA components of latent space')
plt.legend(handles=legend_elements)
plt.show()
plt.savefig("pca_norm.png")
```

```python
#ok now color label as a function of time
#we can't do current for quenches 1 and 3 because the data are weird
#but for now let's do time

#load time of each event (time is in the image file)
q001_t = np.load('/data0/stephey/processed_data/q001_t_cut.npy')
q003_t = np.load('/data0/stephey/processed_data/q003_t_cut.npy')
q076_t = np.load('/data0/stephey/processed_data/q076_t_cut.npy')
q103_t = np.load('/data0/stephey/processed_data/q103_t_cut.npy')

print(q001_t.shape)
```

```python
#see if we can normalize colormap to quench time
#current at quench
q001_q_a = 9677
q003_q_a = 10197
q076_q_a = 15163
q103_q_a = 15986

#time of each quench
q001_q_t = 185911105.0
q003_q_t = 327732493.0
q076_q_t = 176925635.0
q103_q_t = 202845394.0

#normalize time array by quench time, 1 should be quench time
q001_q_norm = q001_t / q001_q_t
q003_q_norm = q003_t / q003_q_t
q076_q_norm = q076_t / q076_q_t
q103_q_norm = q103_t / q103_q_t 
```

```python
#create color map array based on these times
msize = 18

q001_pca_0 = list()
q001_pca_1 = list()
#do this part in advance
for i in range(qall_data.shape[0]):
    if color_label[i] == 'red':
        q001_pca_0.append(img_pca[i,0])
        q001_pca_1.append(img_pca[i,1])     

#now we can plot
plt.figure(figsize=(12, 7))
plt.scatter(q001_pca_0, q001_pca_1, c=q001_q_norm, s=msize)
plt.title('Quench 001 PCA components')
cbar = plt.colorbar()
cbar.set_label('Normalized event time', rotation=90, horizontalalignment='center')
plt.show()
```

```python
#for quench 001 plot images from each cluster
num_images = 3

top_cluster_001 = list()
#find points in the right pca space
for i in range(qall_data.shape[0]):
    if color_label[i] == 'red':
        pca_0 = img_pca[i,0]
        pca_1 = img_pca[i,1]
        if (-30 < pca_0 < -20) and (30 < pca_1 < 40):
            top_cluster_001.append(i)
        
print("These are the images in the top cluster of quench 001:")        
print(top_cluster_001)        

plt.figure(figsize=(30, 10))

#randomly pick three images to show (not the first three)
pick_3 = np.random.choice(top_cluster_001, num_images)

print("These are the random images we chose to display:")
print(pick_3)

show_images = qall_data[pick_3,:,:]

for i, image_idx in enumerate(show_images):
    # plot original image
    ax = plt.subplot(3, num_images, i + 1)
    plt.imshow(image_idx)
    #plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    plt.axis('equal')
    
plt.show()
```

```python
#for quench 001 plot images from each cluster
num_images = 3

bottom_cluster_001 = list()
#find points in the right pca space
for i in range(qall_data.shape[0]):
    if color_label[i] == 'red':
        pca_0 = img_pca[i,0]
        pca_1 = img_pca[i,1]
        if (-20 < pca_0 < 0) and (-20 < pca_1 < -10):
            bottom_cluster_001.append(i)
        
print("These are the images in the bottom cluster of quench 001:")        
print(bottom_cluster_001)        

plt.figure(figsize=(30, 10))

#randomly pick three images to show (not the first three)
pick_3 = np.random.choice(bottom_cluster_001, num_images)

print("These are the random images we chose to display:")
print(pick_3)

show_images = qall_data[pick_3,:,:]

for i, image_idx in enumerate(show_images):
    # plot original image
    ax = plt.subplot(3, num_images, i + 1)
    plt.imshow(image_idx)
    #plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    plt.axis('equal')
    
plt.show()
```
